Задание:

```
<html>
	<head>
		<meta charset="utf-8">
		<title>Форма для заполнения</title>
	</head>
	<body>
		<form action="http://httpbin.org/post" method="POST" 	enctype="multipart/form-data"> 
			<p style="color: red ">
			First name:
			<input type="text" name="first name" size="20" maxlength="50">
			</p>
			<p style="color: orange">
			Last name:
			<input type="text" name="lastname" size="20" maxlength="50">
			</p>
			<p style="color: greenyellow ">
			Group:
			<input type="text" name="group" size="20" maxlength="50">
			</p>
			<p style="color: blue  ">
			Message:
			<input type="textarea" name="message" size="20" maxlength="50">
			</p>
			<p style="color: purple ">
			Myimg:
			<input type="file" value="myimg" name="myimg" size="20" maxlength="150">
			</p>
			<input type="submit" value="submit">
		</form>
	</body>
</html>
```

Проанализируем ответ и опишем работу HTTP протокола в данном случае:

Тег FORM устанавливает форму на веб-странице. Форма предназначена для обмена данными между пользователем и сервером. 

Параметр ACTION
Указывает обработчик, к которому обращаются данные формы при их отправке на сервер. В качестве обработчика может выступать CGI-программа или HTML-документ, который включает в себя серверные сценарии (например, Parser). После выполнения обработчиком действий по работе с данными формы он возвращает новый HTML-документ.

Enctype
Определяет способ кодирования данных формы при их отправке на сервер. Обычно устанавливать значение атрибута enctype не требуется, данные вполне правильно понимаются на стороне сервера.
